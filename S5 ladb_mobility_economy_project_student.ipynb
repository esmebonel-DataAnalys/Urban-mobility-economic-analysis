### 2.3 Corregir formatos numéricos y de fecha

**🎯Objetivo:**
Asegurar que las columnas de fechas y valores numéricos estén en formatos correctos para permitir análisis, cálculos y comparaciones precisas.

**Instrucciones:**

- Convierte las columnas de fecha de `traffic` a formato `datetime`. Haz el cambio a prueba de errores.
- En el dataset `eco`, limpia los valores numéricos:
    - En `city_gdp_capita`: elimina separadores de miles (`.`) y reemplaza las comas (`','`) por puntos (`'.'`) antes de convertir a tipo `float`.
    - En `unemployment_pct`: elimina el símbolo de porcentaje (`%`) y reemplaza las comas (`','`) por puntos (`'.'`) antes de convertir a tipo `float`.
    - En `population_m`: reemplaza las comas (`','`) por puntos (`'.'`) antes de convertir a tipo `float`.
- Finalmente, crea una nueva columna llamada `population` multiplicando `population_m` por 1,000,000 para obtener la población total.


<details>
<summary>Haz clic para ver la pista</summary>
para eliminar símbolos, puedes reemplazarlos por un texto vacío.

```python
# Convertir las columnas de traffic a tipo fecha con pd.to_datetime()
traffic['update_time_utc'] = pd.to_datetime(traffic['update_time_utc'], errors='coerce')
#tu código aquí
traffic['update_time_utc_week_ago'] = pd.to_datetime(traffic['update_time_utc_week_ago'], errors='coerce') #tu código aquí

# verificar el cambio
print(f"Valores convertidos exitosamente: {traffic['update_time_utc'].notna().sum()}")
print(f"Valores que no se pudieron convertir: {traffic['update_time_utc'].isna().sum()}")

print(f"Valores convertidos exitosamente: {traffic['update_time_utc_week_ago'].notna().sum()}")
print(f"Valores que no se pudieron convertir: {traffic['update_time_utc_week_ago'].isna().sum()}")
```

```python
# Limpia separadores y convierte columnas numéricas en eco
# eco['city_gdp_capita'] = ....astype(str).str.replace('.', '').str.replace(',', '.').astype(float)
# eco['unemployment_pct'] = ...
eco['city_gdp_capita'] = eco['city_gdp_capita'].astype(str).str.replace('.', '', regex=False).str.replace(',','.').astype(float)   # "1.234,56" → "1234,56"
eco['unemployment_pct']= eco['unemployment_pct'].astype(str).str.replace('%', '', regex=False).str.strip().astype(float)   # "5,3%" → "5,3"                      
# elimina espacios como "5,3 %"
eco['population_m']= eco['population_m'].astype(str).str.replace(',', '.', regex=False).astype(float)  # "3,21" → "3.21"
# eco['population_m'] = ...
# Calcula la población total en unidades absolutas (Multiplica * 1000000)
# eco['population'] = ...
eco['population'] = eco['population_m'] * 1_000_000
# verificar el cambio
eco.info()
eco.head(3)
```

### 5.1 Unir tráfico (tabla principal) con indicadores económicos

### 6.1 Visualizar relaciones entre economía y tráfico

**🎯Objetivo:**
Analizar visualmente la distribución y la relación entre indicadores de tráfico y economía en 2024, para identificar posibles patrones o tendencias generales entre ambas variables.

**Instrucciones**
- Usa las librerías `seaborn` y `matplotlib.pyplot` para generar los gráficos.
- Visualiza la distribución del **tráfico** (`jams_delay`) mediante:
    - **Boxplot** → para observar la media, mediana y detectar valores atípicos.
- Visualiza la distribución de la **economía** (`city_gdp_capita`) mediante:
    - **Histograma** → para analizar la forma de la distribución y el valor promedio del PIB per cápita.
- Finalmente, **compara ambas variables**, para observar si existe alguna relación entre ellas, haciendo un solo gráfico de barras donde aparezcan ambos indicadores.
- Recuerda agregar título y etiquetas a los ejes de tus gráficos.
- Observa y comenta los patrones, valores extremos o posibles relaciones que identifiques.

**Tip:** Dentro de los parentesis del boxplot, agrega `showmeans=True` para ver la media en el gráfico.

```python

import seaborn as sns
import matplotlib.pyplot as plt

# Filtrar datos de 2024
merged_2024= merged[merged['year'] == 2024]

# Calcular promedio
media_jams = merged_2024['jams_delay'].mean()

# ── 1. Boxplot de jams_delay ──────────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(8, 5))

sns.boxplot(x=merged_2024['jams_delay'], ax=ax, color='steelblue', showmeans=True)

ax.set_title(f'Distribución del Tráfico (jams_delay) — 2024\nPromedio: {media_jams:.2f}', fontsize=14)
ax.set_xlabel('Retraso por tráfico (jams_delay)')
plt.tight_layout()
plt.show()
```

<div class="alert alert-block alert-success"> 
<b>Comentario del Revisor #1</b> <a class="tocSkip"></a>

La visualización es adecuada para comparar la distribución de las variables y detectar outliers, rangos y diferencias entre grupos. Utilizar este tipo de gráfico demuestra un enfoque sólido para el análisis exploratorio.

</div>

```python
# Crear histograma para ver la distribución de la economía (city_gdp_capita)
import seaborn as sns
import matplotlib.pyplot as plt

# Filtrar datos de 2024
merged_2024 = merged[merged['year'] == 2024]

# Calcular promedio
media_gdp = merged_2024['city_gdp_capita'].mean()

# ── Histograma: distribución de city_gdp_capita ───────────────────────────────
fig, ax = plt.subplots(figsize=(8, 5))

sns.histplot(merged_2024['city_gdp_capita'], bins=20, kde=True, color='seagreen', ax=ax)

ax.axvline(media_gdp, color='red', linestyle='--', linewidth=1.5,
           label=f'Media: {media_gdp:,.0f}')

ax.set_title(f'Distribución del PIB per cápita — 2024\nPromedio: {media_gdp:,.0f}', fontsize=14)
ax.set_xlabel('PIB per cápita (city_gdp_capita)')
ax.set_ylabel('Frecuencia')
ax.legend()

plt.tight_layout()
plt.show()
```

<div class="alert alert-block alert-success"> 
<b>Comentario del Revisor #1</b> <a class="tocSkip"></a>

Generaste correctamente el histograma de PIB per cápita y añadiste etiquetas adecuadas. Esta visualización es útil para entender la distribución de la variable, detectar sesgos y evaluar la presencia de valores atípicos.

</div>

```python
# Gráfico de barras para comparar jams_delay y city_gdp_capita por ciudad
#merged.plot( ... , y=['jams_delay', 'city_gdp_capita'])
# Gráfico de barras con doble eje Y
fig, ax1 = plt.subplots(figsize=(14, 6))
ax2 = ax1.twinx()

merged.plot(kind='bar', x='city', y='jams_delay',
            ax=ax1, color='steelblue', alpha=0.7, label='Tráfico (jams_delay)')

merged.plot(kind='bar', x='city', y='city_gdp_capita',
            ax=ax2, color='seagreen', alpha=0.5, label='PIB per cápita')

ax1.set_title('Tráfico vs PIB per cápita por ciudad — 2024', fontsize=14)
ax1.set_xlabel('Ciudad')
ax1.set_ylabel('Retraso por tráfico (jams_delay)', color='steelblue')
ax2.set_ylabel('PIB per cápita (city_gdp_capita)', color='seagreen')

ax1.tick_params(axis='x', rotation=45)
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.tight_layout()
plt.show()
```
